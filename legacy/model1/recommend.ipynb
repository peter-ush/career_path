```python
import json
import os

def add_skill_vector(new_data, path="skill_db.json"):
    # 파일이 없으면 새로 생성
    if not os.path.exists(path):
        skill_db = []
    else:
        with open(path, "r", encoding="utf-8") as f:
            skill_db = json.load(f)

    # 새로운 공고 데이터 추가
    skill_db.append(new_data)

    # 다시 저장
    with open(path, "w", encoding="utf-8") as f:
        json.dump(skill_db, f, ensure_ascii=False, indent=2)

    print(f"{new_data['job_id']} 스킬 벡터 저장 완료!")
```

```python
import json
from openai import OpenAI


def analyze_resume_to_skill_vector(resume_text: str, skill_vocab):
    """
    resume_text: 자기소개서 원문
    skill_vocab: ["Java", "Spring MVC", ...] 형태의 전체 스킬 목록
    return: { "Java": 2, "Spring MVC": 0, ... } 형태 딕셔너리
    """
    skills_str = ", ".join(skill_vocab)

    system_msg = (
        "너는 채용 담당자다. 주어진 자기소개서를 읽고, "
        "각 스킬에 대해 0~3 점수를 매겨라.\n"
        "- 0: 거의 언급/경험 없음\n"
        "- 1: 기초 수준 (수업, 간단 프로젝트 등)\n"
        "- 2: 중급 수준 (여러 프로젝트, 실무 인턴 등)\n"
        "- 3: 고급 수준 (심화 경험, 리드 경험 등)\n"
        "반드시 JSON만 출력해라. 형식은 {\"skills\": {\"스킬명\": 점수, ...}}\n"
    )

    user_msg = (
        f"[스킬 목록]\n{skills_str}\n\n"
        "[자기소개서]\n"
        f"{resume_text}\n"
    )

    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {"role": "system", "content": system_msg},
            {"role": "user", "content": user_msg},
        ],
        temperature=0.0,
    )

    raw = response.choices[0].message.content
    try:
        data = json.loads(raw)
    except json.JSONDecodeError:
        # 혹시 모델이 말 조금 덧붙였을 경우를 대비해 재시도하거나, 간단 후처리 로직 추가 가능
        print("JSON 파싱 실패, 응답:", raw)
        raise

    # 전체 vocab 기준으로 빠진 스킬은 0으로 채워줌
    user_skill_vec = {s: 0 for s in skill_vocab}
    for s, v in data.get("skills", {}).items():
        if s in user_skill_vec:
            user_skill_vec[s] = int(v)

    return user_skill_vec
```

```python
def skill_match_score(job_skill_vec, user_skill_vec):
    numerator = 0
    denominator = 0

    for s, job_score in job_skill_vec.items():
        user_score = user_skill_vec.get(s, 0)

        numerator += min(job_score, user_score)
        denominator += job_score

    if denominator == 0:
        return 0.0

    return numerator / denominator
```

```python
import json
import numpy as np
from chromadb import Client
from openai import OpenAI

class RecommendationEngine:
    def __init__(self, chroma_collection_name="job_rag", skill_db_path="skill_db.json"):
        self.client = OpenAI(api_key="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")    
        self.chroma = Client()
        chroma = Client()

        self.collection = self.chroma.get_or_create_collection(name=chroma_collection_name)

        # 스킬 DB 로드
        with open(skill_db_path, "r", encoding="utf-8") as f:
            self.skill_db = json.load(f)

        # 스킬 vocab 생성
        self.skill_vocab = self._build_skill_vocab()
    def generate_explanation(self, recommendations, top_n=5):
        """
        recommendations = engine.recommend() 결과 리스트
        GPT가 이를 기반으로 사용자에게 설명을 생성
        """

        # 상위 N개만 사용
        recs = recommendations[:top_n]

        # GPT에게 전달할 텍스트 생성
        summary_input = "다음은 추천 결과입니다:\n\n"
        for r in recs:
            summary_input += (
                f"- {r['job_id']}: "
                f"final={r['normalized_score']:.3f}, "
                f"skill={r['skill_score']:.3f}, "
                f"rag={r['rag_score']:.3f}\n"
            )

        summary_input += (
            "\n이 결과를 바탕으로, 사용자의 자기소개서와 각 포지션의 적합도를 "
            "대화형식으로 자연스럽게 설명해줘. "
            "스킬 점수가 높은 이유, RAG 점수가 높은 이유도 함께 말해줘. "
            "사용자가 이해하기 쉽게, 과하지 않게 설명해줘."
        )

        response = self.client.chat.completions.create(
            model="gpt-4.1",   # 4.1-mini도 가능
            messages=[
                {"role": "system", "content": "너는 취업 코치이며 사용자에게 결과를 친절하게 설명한다."},
                {"role": "user", "content": summary_input},
            ],
            temperature=0.4,
        )
        return response.choices[0].message.content

    # ---------------------------
    # 유틸 함수들
    # ---------------------------
    def _build_skill_vocab(self):
        vocab = set()
        for job in self.skill_db:
            for s in job["skills"].keys():
                vocab.add(s)
        return sorted(vocab)

    def embed(self, text):
        emb = self.client.embeddings.create(
            model="text-embedding-3-large",
            input=text
        ).data[0].embedding
        return np.array(emb)

    def cosine_similarity(self, a, b):
        return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

    # ---------------------------
    # 1) 자기소개서 → 스킬 벡터 분석
    # ---------------------------
    def analyze_resume_skills(self, resume_text):
        skills_str = ", ".join(self.skill_vocab)

        system_msg = (
            "너는 채용 담당자다. 주어진 자기소개서를 읽고, "
            "각 스킬에 0~3 점수를 매겨라.\n"
            "- 0: 없음\n- 1: 기초\n- 2: 중급\n- 3: 고급\n"
            "반드시 JSON {\"skills\": {\"스킬명\": 점수}} 형태로만 출력."
        )

        user_msg = f"[스킬 목록]\n{skills_str}\n\n[자기소개서]\n{resume_text}"

        response = self.client.chat.completions.create(
            model="gpt-4.1-mini",
            messages=[
                {"role": "system", "content": system_msg},
                {"role": "user", "content": user_msg},
            ],
            temperature=0.0,
        )

        raw = response.choices[0].message.content
        data = json.loads(raw)

        # 빠진 스킬은 0으로
        vector = {s: 0 for s in self.skill_vocab}
        for s, v in data.get("skills", {}).items():
            if s in vector:
                vector[s] = int(v)
        return vector

    # ---------------------------
    # 2) 스킬 매칭 점수 계산
    # ---------------------------
    def skill_match_score(self, job_skill_vec, user_skill_vec):
        numerator = 0
        denominator = 0
        for s, job_v in job_skill_vec.items():
            user_v = user_skill_vec.get(s, 0)
            numerator += min(job_v, user_v)
            denominator += job_v
        if denominator == 0:
            return 0
        return numerator / denominator

    # ---------------------------
    # 3) RAG 검색
    # ---------------------------
    def search_jobs(self, resume_emb, top_k=5):
        return self.collection.query(
            query_embeddings=[resume_emb],
            n_results=top_k
        )

    # ---------------------------
    # 4) 최종 추천 계산
    # ---------------------------
    def recommend(self, resume_text, top_k=5, alpha=0.6, rag_scale=3.5):
        # 1) 사용자 스킬 벡터
        user_skill_vec = self.analyze_resume_skills(resume_text)

        # 2) 자기소개서 임베딩
        resume_emb = self.embed(resume_text)

        # 3) RAG 검색
        results = self.search_jobs(resume_emb, top_k=top_k)

        job_ids = results["ids"][0]
        job_texts = results["documents"][0]
        distances = results["distances"][0]

        final_results = []

        for i, job_id in enumerate(job_ids):
            job_doc = job_texts[i]
            dist = distances[i]

            # RAGScore = 1 - distance
            rag_score = 1 - dist
            rag_score_scaled = rag_score * rag_scale

            # 스킬 스코어 로드
            job_skill_vec = next(item["skills"] for item in self.skill_db if item["job_id"] == job_id)
            skill_score = self.skill_match_score(job_skill_vec, user_skill_vec)

            # 최종 점수
            final_score = alpha * skill_score + (1 - alpha) * rag_score_scaled

            final_results.append({
                "job_id": job_id,
                "job_text": job_doc,
                "final_score": final_score,
                "skill_score": skill_score,
                "rag_score": rag_score,
                "rag_scaled": rag_score_scaled
            })

        # ---------------------------
        # 5) Min-Max Normalization
        # ---------------------------
        scores = np.array([x["final_score"] for x in final_results])
        min_s, max_s = scores.min(), scores.max()
        norm_scores = (scores - min_s) / (max_s - min_s + 1e-8)

        for x, n in zip(final_results, norm_scores):
            x["normalized_score"] = float(n)

        # 높은 점수 순으로 정렬
        final_results.sort(key=lambda x: x["normalized_score"], reverse=True)

        return final_results

    # 바플롯
    def bar_plot(self, recommendations):
        import matplotlib.pyplot as plt

        job_ids = [r["job_id"] for r in recommendations]
        skill_scores = [r["skill_score"] for r in recommendations]
        rag_scores = [r["rag_score"] for r in recommendations]

        x = np.arange(len(job_ids))
        width = 0.35

        fig, ax = plt.subplots(figsize=(10, 6))
        bars1 = ax.bar(x - width/2, skill_scores, width, label='Skill Score')
        bars2 = ax.bar(x + width/2, rag_scores, width, label='RAG Score')

        ax.set_xlabel('Job ID')
        ax.set_ylabel('Scores')
        ax.set_title('Skill Score vs RAG Score by Job')
        ax.set_xticks(x)
        ax.set_xticklabels(job_ids, rotation=45, ha='right')
        ax.legend()

        plt.tight_layout()
        plt.show()
```
